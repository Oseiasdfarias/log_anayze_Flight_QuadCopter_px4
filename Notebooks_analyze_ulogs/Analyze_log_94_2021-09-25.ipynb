## Analise dos Dados do Magnetômetros

```python
tempo_mag_0 = df_mag_0["timestamp"].values
```

```python
tempo_mag_1 = df_mag_1["timestamp"].values
```


 #### **Dados dos eixos x, y e z dos Magnetômetros**

<br/>

```python
eixo_x_mag_0 = df_mag_0["x"].values
eixo_y_mag_0 = df_mag_0["y"].values
eixo_z_mag_0 = df_mag_0["z"].values
```

```python
eixo_x_mag_1 = df_mag_1["x"].values
eixo_y_mag_1 = df_mag_1["y"].values
eixo_z_mag_1 = df_mag_1["z"].values
```

## **Plotando os Gráficos dos Eixos dos Magnetômetros e Comparando os Valores**

## Eixo ( X ) Mag_0 e Mag_1

```python
plt.figure(figsize = (14, 6))
plt.plot(tempo_mag_0, eixo_x_mag_0, color = "b", label="Medições Mag_0")
plt.plot(tempo_mag_1, eixo_x_mag_1, color = "r", label="Medições Mag_1")

plt.xlabel("Tempo (clock cpu)", fontsize = 15)
plt.ylabel("Medições Eixo ( x ) Mags", fontsize = 15)

plt.legend()
plt.show()
```

## Eixo ( Y ) Mag_0 e Mag_1

```python
plt.figure(figsize = (14, 6))
plt.plot(tempo_mag_0, eixo_y_mag_0, color = "b", label="Medições Mag_0")
plt.plot(tempo_mag_1, eixo_y_mag_1, color = "r", label="Medições Mag_1")

plt.xlabel("Tempo (clock cpu)", fontsize = 15)
plt.ylabel("Medições Eixo ( y ) Mags", fontsize = 15)

plt.legend()
plt.show()
```

## Eixo ( Z ) Mag_0 e Mag_1

```python
plt.figure(figsize = (14, 6))
plt.plot(tempo_mag_0, eixo_z_mag_0, color = "b", label="Medições Mag_0")
plt.plot(tempo_mag_1, eixo_z_mag_1, color = "r", label="Medições Mag_1")

plt.xlabel("Tempo (clock cpu)", fontsize = 15)
plt.ylabel("Medições Eixo ( z ) Mags", fontsize = 15)

plt.legend()
plt.show()
```
